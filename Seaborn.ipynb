# Fundemantals of Seaborn

Seaborn is helpful in statistical related graph and visualization.

We are going to use to examine a **Statistical Estimation** and **Error Bars** example , from the Seaborn library's built-in **fMRI** data. 

```python
# Import Necessary Library

import seaborn as sns
```

```python
# By default, seaborn plots will be made with the current values of the matplotlib rcParams:

sns.barplot(x=["A","B","C"],y=[1,3,2])
```

```python
# You can set a default theme using set_theme() method. This provides a pleasing theme for visualization.

sns.set_theme()
sns.barplot(x=["A","B","C"],y=[1,3,2])
```

```python
# Load an example dataset from the online repository, Use get_dataset_names() to see a list of available datasets.
    
sns.get_dataset_names()
```

```python
# Use the load_dataset() method to load fMRI.

fmri = sns.load_dataset("fmri")
```

fMRI data is a dataframe with 1064 and 5 columns.

Column names; 
* timepoint
* event
* region
* signal

```python
# You can query and inspect data before plotting and visualizing it. 
# Inspect the data with functions such as head() , tail() to view the top and bottom parts of the dataset.

fmri.head()
```

```python
fmri.tail()
```

```python
# You can use the relplot() method to plot graphs.
# Provide necessary parameters like data,x,y,hue,style,size e.t.c

sns.relplot(data=fmri,kind="line",x="timepoint",y="signal",col="region",hue="event",style="event") 
        # Seaborn provides default values for many but the dataset is must.
```

## Analysis Using Seaborn Plots

* **col:** You will notice that there are two graphs. This is because we use **columns** as region and two different colored lines represent stim and cue which are events.
* **x , y :** For both plots, time point(x=axis) and signal (y=axis) are the same.

We infer that signals on the parietal are higher than those on the frontal. 

Stim line shows that it peaks or troughs around the timepoint value of around 5.0 . Similar inferences can be drawn on event cue.

Both the signals are banded.This indicates that Seaborn has already performed statistical estimation and error bars,superimposed on the signals lines.
* Error bars represent the uncertainty of the estimate.

**Many seaborn function automatically  perform the necessary statistical estimation.**

Uses bootstrapping to compute confidence intervals and draw error bars.

This feature of seaborn goes beyond descriptive statistic.

```python
# For example , We can enhances a scatter plot by including a lineer regression model (and its uncertainty) using lmplot()

sns.lmplot(data=fmri,x="timepoint",y="signal",col="region",hue="event")
```

```python
# Seaborn provides the displot() method, using which statistical distribution is automatically plotted and visualized.
# This is impossible another plotting libraries.

sns.displot(data=fmri,kind='hist') # This dataset is not good example to see the power of distplot() function.
```
